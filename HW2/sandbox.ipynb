```python
from gensim import downloader
def data_loader(file_path, embedding_path):
    words = []
    labels = []
    sentences = []
    curr_sentence = []
    curr_sentence_labels = []
    embedding_model = downloader.load(embedding_path)
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    for line in lines:
        if line == "\n" or line == "\t\n":
            sentences.append(curr_sentence)
            labels.append(curr_sentence_labels)
            curr_sentence = []
            curr_sentence_labels = []
        else:
            curr_word, curr_label = line.split('\t') 
            coded_label = 0 if curr_label == 'O' else 1
            curr_sentence_labels.append(coded_label)
            curr_sentence.append(curr_word)
            words.append(curr_word)
    
    return sentences, labels, words
```

```python
from data_load import data_loader
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score


TRAIN_PATH = "data/train.tagged"
TEST_PATH = "data/dev.tagged"
EMBEDDING_PATH = 'word2vec-google-news-300'

train_embedded_words, train_words_labels = data_loader(TRAIN_PATH, EMBEDDING_PATH)
print("finished load train")
test_embedded_words, test_words_labels = data_loader(TEST_PATH, EMBEDDING_PATH)
print("finished load test")
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(train_embedded_words, train_words_labels)
```
